```python
train_data = pd.read_csv("//kaggle/input/teco-psda-exercisesheet1-hdma-2024/hdma_train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("/kaggle/input/teco-psda-exercisesheet1-hdma-2024/hdma_test.csv")
test_data.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, accuracy_score



train_data.fillna(train_data.mode().iloc[0], inplace=True)
train_encoded = pd.get_dummies(train_data.drop(['Unnamed: 0', 'deny'], axis=1), drop_first=True)
X_train = train_encoded
y_train = train_data['deny']


rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)


cv_scores = cross_val_score(rf_classifier, X_train, y_train, cv=5)


test_data.fillna(test_data.mode().iloc[0], inplace=True)
test_encoded = pd.get_dummies(test_data.drop('Unnamed: 0', axis=1), drop_first=True)
X_test = test_encoded[X_train.columns] 



y_pred = rf_classifier.predict(X_test)
print(sum(y_pred)/len(y_pred))
output = pd.DataFrame({'Id': test_data.iloc[:,0], 'Expected': y_pred})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
